```python
import os
from io import BytesIO
import zipfile

output_dir = "my_tables"
os.makedirs(output_dir, exist_ok=True)

with zipfile.ZipFile(BytesIO(r.content)) as zf:
    for i, name in enumerate(zf.namelist()):
        if name.endswith('.csv'):
            content = zf.read(name)
            # Генерируем простое имя: table_1.csv, table_2.csv
            simple_name = f"table_{i+1}.csv"
            with open(os.path.join(output_dir, simple_name), "wb") as f:
                f.write(content)
            print(f"Сохранён: {simple_name} ({len(content)} байт)")
```
